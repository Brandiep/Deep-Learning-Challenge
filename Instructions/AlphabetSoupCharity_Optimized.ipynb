## Preprocessing

```python

# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
from pathlib import Path
```

```python
application_df = pd.read_csv(Path("/content/drive/MyDrive/Deep-Learning-Challenge/Instructions/charity_data.csv"))
application_df.head()

```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(["EIN", "NAME"], axis='columns')
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_counts= application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(application_counts[application_counts < 500].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#application_df['CLASSIFICATION'].value_counts() >1
application_df['CLASSIFICATION'].value_counts().loc[lambda x : x>1] 
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classification_counts= application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = list(classification_counts[classification_counts< 1000].index)
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
clean_app_df = pd.get_dummies(application_df)
clean_app_df
```

```python
# Split our preprocessed data into our features and target arrays
X = clean_app_df.drop('IS_SUCCESSFUL',1)
y = clean_app_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a MinMaxScaler instances
scaler = MinMaxScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_train_scaled
```

## Compile, Train and Evaluate the Model

Optimization Attempts

```python
#Attempt 1 - add another hidden layer

# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn1 = tf.keras.models.Sequential()
input_dimentions = len(X_train_scaled[0])

# First hidden layer
nn1.add(tf.keras.layers.Dense(units=10, activation="relu", input_dim=input_dimentions))

# Second hidden layer
nn1.add(tf.keras.layers.Dense(units=6, activation="relu"))

#Third hidden layer
nn1.add(tf.keras.layers.Dense(units=4, activation="relu"))

# Output layer
nn1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn1.summary()
```

```python
# Compile the model
nn1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
2# Train the model
fit_model1 = nn1.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss1, model_accuracy1 = nn1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss1}, Accuracy: {model_accuracy1}")
```

Adding an additional layer resulted in slightly increased accuracy

```python
#Attempt 2 - use a different activation function + a third layer

# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn2 = tf.keras.models.Sequential()
input_dimentions = len(X_train_scaled[0])

# First hidden layer
nn2.add(tf.keras.layers.Dense(units=10, activation="tanh", input_dim=input_dimentions))

# Second hidden layer
nn2.add(tf.keras.layers.Dense(units=6, activation="tanh"))

#Third hidden layer
nn2.add(tf.keras.layers.Dense(units=4, activation="tanh"))

# Output layer
nn2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn2.summary()
```

```python
# Compile the model
nn2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
2# Train the model
fit_model2 = nn2.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss2, model_accuracy2 = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss2}, Accuracy: {model_accuracy2}")
```

changing the activation function increased the accuracy of the model a little more
